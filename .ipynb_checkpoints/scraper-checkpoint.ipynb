```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
class PLAYER:
    def __init__(self, baseurl):
        self.baseurl = baseurl
    def player_url(self,first,last):
        purl = self.baseurl.format(first[0],last[:4],first[:2])
        return purl
class SCRAPER(PLAYER):
    def __init__(self, baseurl):
        PLAYER.__init__(self, baseurl)
    def get_player_stats(self):
        PLAYER_URL = self.baseurl
        page = requests.get(PLAYER_URL)
        soup = BeautifulSoup(page.text, "html.parser")
        table = soup.find("table")
        df = pd.read_html(str(table))[0]
        df.rename(columns={"Unnamed: 7_level_1":"Home"}, inplace=True)
        renamed_cols = [' '.join(col).strip() if "Unnamed" not in col[0] else col[1] for col in df.columns.values]
        df.columns = renamed_cols
        # if home == 0 then false (i.e. away) elif home == 1 then true
        df.Home = np.where(df['Home'] == "@", 0, 1)
        return df
        
```

```python
url = PLAYER("https://www.pro-football-reference.com/players/{}/{}{}00/gamelog/")
```

```python
player = url.player_url("Tom","Brady")
```

```python
SCRAPER(player).get_player_stats()
```

```python
def scrape(first,last):
    PLAYER_URL =f"https://www.pro-football-reference.com/players/{first[0]}/{last[:4]}{first[:2]}00/gamelog/"
    page = requests.get(PLAYER_URL)
    soup = BeautifulSoup(page.text, "html.parser")
    table = soup.find("table")
    df = pd.read_html(str(table))[0]
    df.rename(columns={"Unnamed: 7_level_1":"Home"}, inplace=True)
    renamed_cols = [' '.join(col).strip() if "Unnamed" not in col[0] else col[1] for col in df.columns.values]
    df.columns = renamed_cols
    # if home == 0 then false (i.e. away) elif home == 1 then true
    df.Home = np.where(df['Home'] == "@", 0, 1)
    return df
```

```python
def create_table(cursor = c):
    cursor.execute(''' CREATE TABLE stats(Rk INT, Year INT, Game_num INT, Week INT, Age REAL, Team TEXT, Home INT, 
                  Opp_team TEXT, Result TEXT, Game_start TEXT, Pass_cmp INT, Pass_atmpt INT, Pass_perc_cmp REAL,
                  Pass_yds_gained INT, Pass_TD INT, Pass_intcpt INT, Pass_yds_lost INT, Pass_ypa REAL, Pass_aypa REAL,
                  Rush_atmpt INT, Rush_yds_gained INT, Rush_ypa REAL, Rush_TD INT, 
                  Receive_tgt INT, Receive_rec INT, Receive_yds INT, Receive_ypr REAL, Receive_TD INT, 
                  Receive_catchp REAL, Receive_ypt REAL)
    ''')
```

```python
def populate_table(table_name:str, df:pd.DataFrame):
    rows = [tuple(df.iloc[n].values) for n in range(len(df))]
    c.executemany(f'''INSERT INTO {table_name} VALUES({','.join(len(df.columns)*('?'))})''', rows)
    conn.commit()
```

```python
populate_table('stats',df)
```

```python
try:
    sqliteConnection = sqlite3.connect('SQLite_Python.db')
    cursor = sqliteConnection.cursor()
    print("Database created and Successfully Connected to SQLite")

    sqlite_select_Query = "select sqlite_version();"
    cursor.execute(sqlite_select_Query)
    record = cursor.fetchall()
    print("SQLite Database Version is: ", record)
    cursor.close()

except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if sqliteConnection:
        sqliteConnection.close()
        print("The SQLite connection is closed")
```
