```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
class PLAYER:
    def __init__(self, baseurl):
        self.baseurl = baseurl
    def player_url(self,first,last):
        purl = self.baseurl.format(first[0],last[:4],first[:2])
        return purl
class SCRAPER(PLAYER):
    def __init__(self, baseurl):
        PLAYER.__init__(self, baseurl)
    def get_player_stats(self):
        PLAYER_URL = self.baseurl
        page = requests.get(PLAYER_URL)
        soup = BeautifulSoup(page.text, "html.parser")
        table = soup.find("table")
        df = pd.read_html(str(table))[0]
        df.rename(columns={"Unnamed: 7_level_1":"Home"}, inplace=True)
        renamed_cols = [' '.join(col).strip() if "Unnamed" not in col[0] else col[1] for col in df.columns.values]
        df.columns = renamed_cols
        # if home == 0 then false (i.e. away) elif home == 1 then true
        df.Home = np.where(df['Home'] == "@", 0, 1)
        return df
        
```

```python
url = PLAYER("https://www.pro-football-reference.com/players/{}/{}{}00/gamelog/")
```

```python
player = url.player_url("Tom","Brady")
```

```python
SCRAPER(player).get_player_stats()
```
